```python
# Dependencies
import pandas as pd
```

```python
# List of file names to be combined 
files21 = ['citbike\JC-202101-citibike-tripdata.csv']

# Create an empty list to store each dataframe
df21_list = []

# Loop through each file in the list
for file in files21:
    # Read the file into a dataframe
    df21 = pd.read_csv(file)

    # Append the dataframe to the list
    df21_list.append(df21)

# Concatenate all dataframes in the list into one dataframe
df21 = pd.concat(df21_list)

# Show dataframe
df21
```

```python
# List of file names to be combined 
files22 = ['citbike\JC-202102-citibike-tripdata.csv']

# Create an empty list to store each dataframe
df22_list = []

# Loop through each file in the list
for f in files22:
    # Read the file into a dataframe
    df22 = pd.read_csv(f)

    # Append the dataframe to the list
    df22_list.append(df22)

# Concatenate all dataframes in the list into one DataFrame
df22 = pd.concat(df22_list)

# Show dataframe
df22
```

```python
# Concatenate all dataframes for all the years into one
combined_df = pd.concat([df21, df22,], axis=0)

# Display the results
combined_df
```

```python
# Dictionary mapping values in the 'gender' column to new values
mapping = {0: 'Unknown', 1: 'Male', 2: 'Female'}

# Use the 'map' method to change the values in the 'gender' column
combined_df['gender'] = combined_df['gender'].map(mapping)

# Show DataFrame
combined_df
```

```python
# Save the dataframe to a CSV file
combined_df.to_csv('datecitibike.csv', index=False)
```
